```python
#3.Generate and Visualize a Random Matrix:

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# step 1: 4x4 random matrix making (float values: 0 to 1 )
matrix = np.random.rand(4, 4)

# step 2: Heatmap plot making
plt.figure(figsize=(6, 5))

# Heatmap draw
sns.heatmap(matrix, annot=True, cmap="YlGnBu", fmt=".2f",
            xticklabels=[f'Col {i+1}' for i in range(4)],
            yticklabels=[f'Row {i+1}' for i in range(4)])

# title and layout
plt.title("Heatmap of 4x4 Random Matrix")
plt.tight_layout()

# show
plt.show()

```
